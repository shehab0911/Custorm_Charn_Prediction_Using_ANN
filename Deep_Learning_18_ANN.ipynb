```python
yes_no_column=['Partner','Dependents','tenure','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_column:
  df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.columns
```

```python
df2.sample(4)
```

```python
df2.dtypes
```

```python
cols_to_scale=['tenure','MonthlyCharges','TotalCharges']

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()

df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.sample(3)
```

```python
for col in df2:
  print(f'{col}:{df2[col].unique()}')
```

```python
x=df2.drop('Churn',axis='columns')
y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
x_test[:10]
```

```python
x_train[:10]
```

```python
len(x_train.columns)
```

```python
from keras.engine.training import optimizer
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')

])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp=model.predict(x_test)
yp[:5]
```

```python
y_test[:5]
```

```python
y_pred=[]
for element in yp:
  if element>0.5:
    y_pred.append(1)
  else:
    y_pred.append(0)
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

```python
import seaborn as sn
cm= tf.math.confusion_matrix(labels=y_test,predictions=y_pred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
